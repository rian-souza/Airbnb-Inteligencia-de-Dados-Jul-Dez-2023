```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("/content/Airbnb_site_hotel new.csv", dtype=str)

df['price'] = df['price'].str.replace(',', '.').str.replace('[^0-9.]', '', regex=True).astype(float)

df["host Certification"] = pd.to_numeric(df["host Certification"], errors='coerce')
df["host Certification"] = df["host Certification"].fillna(df["host Certification"].mean())
df["room_type"] = df["room_type"].fillna(df["room_type"].mode()[0])
df["host total listings count"] = df["host total listings count"].fillna(df["host total listings count"].mode()[0])
df["consumer"] = df["consumer"].fillna(df["consumer"].mode()[0])

df["host response rate"] = df["host response rate"].str.replace(",", ".").str.rstrip("%").astype(float)
df["host acceptance rate"] = df["host acceptance rate"].str.replace(",", ".").str.rstrip("%").astype(float)

df["price"] = df["price"].astype(float)

df["price"].hist(bins=30)
plt.title("Distribuição de Preços")
plt.xlabel("Preço")
plt.ylabel("Frequência")
plt.show()

df["reply time"] = df["reply time"].fillna(df["reply time"].mode()[0])

mediana_price = df["price"].median()
df["price"] = df["price"].fillna(mediana_price)

df.isnull().sum()

```
